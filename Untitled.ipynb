```python
# coding: utf-8
import matplotlib.pyplot as plt
import numpy as np
import math
epsilon = 0.0025
delta = 0.05
big = 1

# f(x)を自分で書く
f = lambda x: big * (math.exp(-x**2/2)) / math.sqrt(2*math.pi)
f_minus = lambda x: big * (math.exp(-(x - epsilon)**2/2)) / math.sqrt(2*math.pi)
f_plus = lambda x: big * (math.exp(-(x + epsilon)**2/2)) / math.sqrt(2*math.pi)
f_Xd = lambda x: big * 0.50 * ((1+delta)*f_minus(x) - (1-delta)*f_plus(x))

# ベクトルxを [-5.0, ..., 5.0] の区間で作成
n = np.linspace(-0.50, 0.50, 10000)

# f(x)の結果を得る
p = []
p_minus = []
p_plus = []
p_Xd = []

for i in range(len(n)):
    p.append(f(n[i]) * 1.41)
    p_minus.append(f_minus(n[i]))
    p_plus.append(f_plus(n[i]))
    p_Xd.append(f_Xd(n[i]))

fig, ax = plt.subplots(1,1)

ax.axvline(0, 0, 1, ls = '--', c = 'red')# グラフに表示
ax.axvline(epsilon / delta, 0, 1, ls = '--')# グラフに表示

# plt.plot(n, p)
# plt.plot(n, p_minus)
# plt.plot(n, p_plus)
plt.plot(n, p_Xd)
plt.show()
```
